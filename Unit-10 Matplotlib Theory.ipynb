```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,8,111,69,103,76,94,78,77,85,55])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
plt.scatter(x,y,c=colors,cmap='viridis_r')#iin exam if cant remeber viridis then write our name and error 
#will show many cmap values
#alpha=0.5 for opacity 0 is transparent and 1 is opaque
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(10,50,size=20)#50 exclude
y=np.random.randint(10,50,size=20)
colors=np.random.randint(10,50,size=20)
sizes=10*np.random.randint(10,50,size=20)
plt.scatter(x,y,c=colors,s=sizes,cmap='nipy_spectral_r')
plt.colorbar()
plt.show()
```

# BAR PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x=np.array(["A","B","C","D"])
y=np.array([3,9,1,10])
color=np.array(['red','blue','green','magenta'])
plt.bar(x,y,color='orange',label="A,B,C,D",width=0.4)

x=np.array(["P","Q","R","S"])
y=np.array([7,5,1,10])
plt.bar(x,y,color='darkblue',label="P,Q,R,S",width=0.5)
plt.legend()
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,2,3,3,3,4,4,5,7,8,8,9,10,10])
plt.hist(x,bins=5)#no two parameters bins divides values in 5 parts
plt.show()
plt.hist(x,bins=[1,5,10])#passing list as parameter in bins so we can modify histogram.
plt.show()
plt.hist(x,orientation='horizontal')
plt.show()
plt.hist(x)
plt.show()
```

# Pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
y=np.array([45,25,25,15])#each piece of pie chart is called wedge.
mylabels=["Apple","Banana","Cherry","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,startangle=90,labels=mylabels,explode=myexplode,shadow=False,autopct="%1.1f%%")
plt.legend()
plt.suptitle("Fruits")
plt.show()


plt.figure(figsize=(10,10))
y=np.array([20,20,20,20,20])
mylabels1=["Dominos","Martinoz","LaPinoz","Uncle Sams's","PizzaHut"]
myexplode1=[0,0,0,0.1,0.45]
mycolors=["LightBlue","indigo","green","Blue","red"]
plt.pie(y,startangle=60,labels=mylabels1,explode=myexplode1,shadow=True,colors=mycolors,autopct="%1.1f%%")
plt.legend()
#plt.title("Pizza")#without space
plt.suptitle("Pizza")#with space
plt.show()
```
